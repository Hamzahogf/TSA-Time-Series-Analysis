```r
# MINI-PROJECT: LOAD CONSUMPTION DATA ALGERIA
# SOURCE: https://data.mendeley.com/datasets/z5x2d3mhw7/1  BY Miss. FARAH RANIA
library(tseries)
# library(dplyr)
library(TSA)
# install.packages("ggplot2", type = "binary")
#install.packages("forecast", type = "binary")
library(forecast)

plotBoxTest <- function(dat, df=1){
  par(mfrow=c(2,1))
  pvs = rep(1, 16)
  for(i in (df+1):16)
    pvs[i] = Box.test(dat, lag=i, fitdf=df)$p.value
  plot(pvs, las=1, log="y", main="Box Pierce")
  abline(h=10^(-(1:10)), col='gray')
  abline(h=0.05, col="red", lty=2)
  pvs = rep(1, 16)
  for(i in (df+1):16)
    pvs[i] = Box.test(dat, lag=i, type="Ljung", fitdf=df)$p.value
  plot(pvs, las=1, log="y", main="Ljung Box")
  abline(h=10^(-(1:10)), col='gray')
  abline(h=0.05, col="red", lty=2)
  par(mfrow=c(1,1))
}
```
