```r
library(TSA)
vent <- ts(vent$vent, start=1)
colnames(vent) <-c('vent')
vent <- ts(vent$vent, start=1, frequency=1)
plot(vent)
```

```r
model <- lm(vent$vent ~ vent$time) 
model <- lm(vent ~ time) 
model <- lm(vent ~ time(vent)) 
summary(model)
plot(vent)
abline(model, col='red')
```

```r
res <- residuals(model)
plot(res)
abline(h=0, col='blue')
```
